```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Chargement du fichier CSV
# Remplace par le bon chemin si nécessaire
fichier = "data1.csv"
df = pd.read_csv(fichier)

print("✅ Aperçu du dataset :")
display(df.tail(10))
```

```python
# %% [4] — Étude de la corrélation
import numpy as np

print("📊 Matrice de corrélation (valeurs numériques seulement) :")
correlation = df.corr(numeric_only=True)
display(correlation)

# Affichage graphique avec seaborn
plt.figure(figsize=(14, 10))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de Corrélation")
plt.show()
```

```python
df['anomaly_type'] = df['anomaly_type'].astype('category').cat.codes
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
import pandas as pd

# 1. Load the data
df = pd.read_csv("data1.csv")

# 2. Drop unnecessary columns (e.g., miner_id)
df.drop(columns=["miner_id"], inplace=True, errors="ignore")

# 3. Force columns to numeric except 'anomaly_type'
for col in df.columns:
    if col != "anomaly_type":
        df[col] = pd.to_numeric(df[col], errors="coerce")

# 4. Remove rows containing NaN values
df.dropna(inplace=True)

# 5. Reset the index
df.reset_index(drop=True, inplace=True)

# 6. Save to a cleaned file
df.to_csv("data1_clean_simple.csv", index=False)
```
