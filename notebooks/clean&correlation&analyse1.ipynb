```python
# %% â€” Simple cleaning of the dataset smart_grid_dataset1.csv

import pandas as pd
import numpy as np

# ðŸ“¥ Load the data
df = pd.read_csv("smart_grid_dataset1_with_energy.csv")

# ðŸ§¼ Basic cleaning
df.columns = df.columns.str.strip()

# ðŸ•’ Convert Timestamp to datetime
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%d-%m-%Y %H.%M", errors='coerce')

# ðŸ”¢ Convert numeric columns
for col in df.columns:
    if col != "Timestamp":
        df[col] = pd.to_numeric(df[col], errors="coerce")

# ðŸ” Missing values before filling
print("\nðŸ” Missing values per column before filling:")
print(df.isnull().sum())

# âž• Fill missing values
df.fillna(method="ffill", inplace=True)

# âœ… Missing values after treatment
print("\nâœ… Missing values per column after filling:")
print(df.isnull().sum())

# ðŸ” Replace Timestamp with an integer of type YYYYMMDDHHMM
df["Timestamp"] = (
    df["Timestamp"].dt.year * 100000000 +
    df["Timestamp"].dt.month * 1000000 +
    df["Timestamp"].dt.day * 10000 +
    df["Timestamp"].dt.hour * 100 +
    df["Timestamp"].dt.minute
).astype("int64")

# ðŸ’¾ Export
output_path = "smart_grid_dataset1_clean.csv"
df.to_csv(output_path, index=False)
print(f"\nðŸ’¾ Cleaned dataset exported to: {output_path}")
```

```python
# %% [1] â€” Nettoyage et prÃ©paration
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Charger les donnÃ©es
df = pd.read_csv("smart_grid_dataset1.csv")

# Supprimer les doublons
df = df.drop_duplicates()

# Convertir la colonne Timestamp en datetime
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%d-%m-%Y %H.%M")

# Extraire des infos utiles de Timestamp (facultatif)
df["Hour"] = df["Timestamp"].dt.hour
df["Day"] = df["Timestamp"].dt.day
df["Month"] = df["Timestamp"].dt.month

# Supprimer Timestamp si tu ne veux pas l'utiliser tel quel
df = df.drop(columns=["Timestamp"])

# SÃ©parer features et cible
X = df.drop(columns=["Overload", "Transformer_Fault"])
y = df["Overload"]  # ou "Transformer_Fault"

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# %% [2] â€” Split des donnÃ©es
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# %% [3] â€” EntraÃ®nement de plusieurs modÃ¨les
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# CrÃ©er les modÃ¨les
models = {
    "RandomForest": RandomForestClassifier(random_state=42),
    "LogisticRegression": LogisticRegression(class_weight='balanced', max_iter=1000),
    "SVM": SVC(class_weight='balanced')
}

# EntraÃ®ner tous les modÃ¨les
fitted_models = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    fitted_models[name] = model
    print(f"âœ… ModÃ¨le entraÃ®nÃ© : {name}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

for name, model in fitted_models.items():
    print(f"\nðŸ“Š RÃ©sultats pour : {name}")
    y_pred = model.predict(X_test)
    
    # Rapport de classification
    print(classification_report(y_test, y_pred, digits=4))
    
    # Matrice de confusion
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Matrice de confusion â€” {name}")
    plt.xlabel("PrÃ©dit")
    plt.ylabel("RÃ©el")
    plt.tight_layout()
    plt.show()
```

```python
# %% [1] â€” Imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# %% [2] â€” Chargement et nettoyage du dataset
df = pd.read_csv("smart_grid_dataset1.csv")

# Supprimer les doublons
df = df.drop_duplicates()

# Convertir la colonne Timestamp
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%d-%m-%Y %H.%M")

# Extraire les infos utiles de Timestamp
df["Hour"] = df["Timestamp"].dt.hour
df["Day"] = df["Timestamp"].dt.day
df["Month"] = df["Timestamp"].dt.month

# Supprimer Timestamp si non utilisÃ©
df = df.drop(columns=["Timestamp"])

# SÃ©parer les features et la cible
X = df.drop(columns=["Overload", "Transformer_Fault"])
y = df["Overload"]

# Normalisation des features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# %% [3] â€” Split train / test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# âœ… Afficher la distribution des classes
print("âœ… Classe dans le train :", dict(pd.Series(y_train).value_counts()))

# %% [4] â€” DÃ©finir et entraÃ®ner les modÃ¨les avec class_weight='balanced'
models = {
    "RandomForest": RandomForestClassifier(random_state=42, class_weight="balanced"),
    "LogisticRegression": LogisticRegression(class_weight="balanced", max_iter=1000),
    "SVM": SVC(class_weight="balanced")
}

fitted_models = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    fitted_models[name] = model
    print(f"âœ… ModÃ¨le entraÃ®nÃ© : {name}")

# %% [5] â€” Ã‰valuation
for name, model in fitted_models.items():
    print(f"\nðŸ“Š RÃ©sultats pour : {name}")
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred, digits=4))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Matrice de confusion â€” {name}")
    plt.xlabel("PrÃ©dit")
    plt.ylabel("RÃ©el")
    plt.tight_layout()
    plt.show()
```
