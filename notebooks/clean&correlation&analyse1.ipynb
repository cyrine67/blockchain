```python
# %% — Simple cleaning of the dataset smart_grid_dataset1.csv

import pandas as pd
import numpy as np

# 📥 Load the data
df = pd.read_csv("smart_grid_dataset1_with_energy.csv")

# 🧼 Basic cleaning
df.columns = df.columns.str.strip()

# 🕒 Convert Timestamp to datetime
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%d-%m-%Y %H.%M", errors='coerce')

# 🔢 Convert numeric columns
for col in df.columns:
    if col != "Timestamp":
        df[col] = pd.to_numeric(df[col], errors="coerce")

# 🔍 Missing values before filling
print("\n🔍 Missing values per column before filling:")
print(df.isnull().sum())

# ➕ Fill missing values
df.fillna(method="ffill", inplace=True)

# ✅ Missing values after treatment
print("\n✅ Missing values per column after filling:")
print(df.isnull().sum())

# 🔁 Replace Timestamp with an integer of type YYYYMMDDHHMM
df["Timestamp"] = (
    df["Timestamp"].dt.year * 100000000 +
    df["Timestamp"].dt.month * 1000000 +
    df["Timestamp"].dt.day * 10000 +
    df["Timestamp"].dt.hour * 100 +
    df["Timestamp"].dt.minute
).astype("int64")

# 💾 Export
output_path = "smart_grid_dataset1_clean.csv"
df.to_csv(output_path, index=False)
print(f"\n💾 Cleaned dataset exported to: {output_path}")
```

```python
# %% [1] — Nettoyage et préparation
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Charger les données
df = pd.read_csv("smart_grid_dataset1.csv")

# Supprimer les doublons
df = df.drop_duplicates()

# Convertir la colonne Timestamp en datetime
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%d-%m-%Y %H.%M")

# Extraire des infos utiles de Timestamp (facultatif)
df["Hour"] = df["Timestamp"].dt.hour
df["Day"] = df["Timestamp"].dt.day
df["Month"] = df["Timestamp"].dt.month

# Supprimer Timestamp si tu ne veux pas l'utiliser tel quel
df = df.drop(columns=["Timestamp"])

# Séparer features et cible
X = df.drop(columns=["Overload", "Transformer_Fault"])
y = df["Overload"]  # ou "Transformer_Fault"

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# %% [2] — Split des données
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# %% [3] — Entraînement de plusieurs modèles
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Créer les modèles
models = {
    "RandomForest": RandomForestClassifier(random_state=42),
    "LogisticRegression": LogisticRegression(class_weight='balanced', max_iter=1000),
    "SVM": SVC(class_weight='balanced')
}

# Entraîner tous les modèles
fitted_models = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    fitted_models[name] = model
    print(f"✅ Modèle entraîné : {name}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

for name, model in fitted_models.items():
    print(f"\n📊 Résultats pour : {name}")
    y_pred = model.predict(X_test)
    
    # Rapport de classification
    print(classification_report(y_test, y_pred, digits=4))
    
    # Matrice de confusion
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Matrice de confusion — {name}")
    plt.xlabel("Prédit")
    plt.ylabel("Réel")
    plt.tight_layout()
    plt.show()
```

```python
# %% [1] — Imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# %% [2] — Chargement et nettoyage du dataset
df = pd.read_csv("smart_grid_dataset1.csv")

# Supprimer les doublons
df = df.drop_duplicates()

# Convertir la colonne Timestamp
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%d-%m-%Y %H.%M")

# Extraire les infos utiles de Timestamp
df["Hour"] = df["Timestamp"].dt.hour
df["Day"] = df["Timestamp"].dt.day
df["Month"] = df["Timestamp"].dt.month

# Supprimer Timestamp si non utilisé
df = df.drop(columns=["Timestamp"])

# Séparer les features et la cible
X = df.drop(columns=["Overload", "Transformer_Fault"])
y = df["Overload"]

# Normalisation des features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# %% [3] — Split train / test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# ✅ Afficher la distribution des classes
print("✅ Classe dans le train :", dict(pd.Series(y_train).value_counts()))

# %% [4] — Définir et entraîner les modèles avec class_weight='balanced'
models = {
    "RandomForest": RandomForestClassifier(random_state=42, class_weight="balanced"),
    "LogisticRegression": LogisticRegression(class_weight="balanced", max_iter=1000),
    "SVM": SVC(class_weight="balanced")
}

fitted_models = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    fitted_models[name] = model
    print(f"✅ Modèle entraîné : {name}")

# %% [5] — Évaluation
for name, model in fitted_models.items():
    print(f"\n📊 Résultats pour : {name}")
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred, digits=4))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Matrice de confusion — {name}")
    plt.xlabel("Prédit")
    plt.ylabel("Réel")
    plt.tight_layout()
    plt.show()
```
